```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Beverage.csv')
df
```

```python
df['volatile acidity'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['volatile acidity'] = le.fit_transform(df['volatile acidity'])

df['volatile acidity'].value_counts()
```

```python
df['citric acid'].value_counts()
```

```python
df['citric acid'] = le.fit_transform(df['citric acid'])
```

```python
df['citric acid'].value_counts()
```

```python
ls = ['fixed acidity','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality']
for col in ls:
  df[col] = le.fit_transform(df[col])

df.info()
```

```python
X = df.drop(columns='volatile acidity')
y = df['volatile acidity']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
print(X.shape)
print(y.shape)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

def LR(X_train, X_test, y_train, y_test):
  LogReg = LogisticRegression()

  LogReg.fit(X_train, y_train)
  y_pred = LogReg.predict(X_test)

  print(confusion_matrix(y_test, y_pred))
  print(classification_report(y_test, y_pred))
```

```python
LR(X_train, X_test, y_train, y_test)
```
